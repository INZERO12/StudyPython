```python
tags_total = []
# 해시태그 단어 하나하나를 모두 추출하여 하나의 리스트로 만들기
for tags in raw_total['tags'] :
    for tag in tags[1:-1].split(',') : 
        tags_total.append(tag[1:].replace('\'', ''))
tags_total
```

```python
tags_total = []
# 해시태그 단어 하나하나를 모두 추출하여 하나의 리스트로 만들기
for tags in raw_total['tags'] :
    for tag in tags[2:-2].split("', '") : 
        tags_total.append(tag.replace('\'', ''))
tags_total
```

```python
from collections import Counter
```

```python
tag_counts = Counter(tags_total)
tag_counts
```

```python
# 가장 많은 count 조회
tag_counts.most_common(50)
```

```python
# 키워드 정제하기
# 정제할 키워드에 대한 리스트 만들기
STOPWORDS = ["#일상", "#선팔", "#제주도", "#jeju", "#반영구",
              "#제주자연눈썹","서귀포눈썹문신","#제주눈썹문신",
              "#소통","#맞팔"]
```

```python
tag_total_selected = []

for tag in tags_total:
    if tag not in STOPWORDS :
        tag_total_selected.append(tag)
tag_total_selected
```

```python
tag_total_selected = Counter(tag_total_selected)
tag_total_selected
```
