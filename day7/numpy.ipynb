## 리스트와 행렬의 차이점
- 행렬구조에 특화된 numpy 모듈에 대하여서 먼저 학습
- 행렬은 반드시 행과 열의 갯수가 인덱싱마다 같아야하고, 연산이 가능함 자료의 성격이 같아야함
- 행렬에 특화된 모듈이 numpy

```python
# 이전에 쓰던 방법
a = [1,2,3,4,5]
b = [10,20,30,40]
val = 3
cc=[]

for x1,x2 in zip(a,b):
    cc.append(a + b)
cc
```

```python
# numpy 구조는 되도록 행과 열의 구조가 같아야한다.
import numpy as np
x = [[2,3,4],[22,33,44]]    # 2행 3열 구조 / 리스트 구조는 브로드캐스트 불가
x2=np.array(x) + 2
x2

len(x)          # 행
len(x[0])       # 열
np.shape(x)     # 2행 3열 구조
np.shape(x2)    # x2는 배열로 변환되었기 때문에 x2.shape()


# aa.bb 구조는 array구조만 가능
```

```python
# 리스트 : 사용자가 막 사용할 수 있도록 만든 구조(입력하기 사용)
# 배열 : 계산에 최적화된 행과 열 처리 기법(속도가 빠름, GPU등, 게임 필수)
# 파이썬에서는 배열에 최적화된 연산식을 Numpy 모듈에서 제공함
# 인공지능에서는 numpy 필수
```

```python
# 1개의 픽셀에 0~255 값이 들어가면 그레이스케일(0은 검정, 255는 흰색)
import matplotlib.pyplot as pit
x=[[0,255,100], [255,0,200]]
pit.imshow(x, cmap='gray')

x2 = np.array(x)
x3=x2-300
print(x3)
pit.imshow(x3, cmap='gray')
```
